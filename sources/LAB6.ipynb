```python
filter = [[0, 1, 0],
          [1, 2, 1],
          [0, 1, 0]]
img = [[1, 8, 6, 6],
       [6, 3, 11, 8],
       [8, 8, 9, 10],
       [9, 10, 10, 7]]
```

```python
def conv(A, Mask):
    res = np.zeros((len(A), len(A[0])))
    for i in range(len(A)):
        for j in range(len(A[i])):
            x = 0
            for i1 in range(-1,2):
                for j1 in range(-1,2):
                    if (i+i1>=0 and i+i1<len(A) and j+j1>=0 and j+j1<len(A[i])):
                        x+=A[i+i1][j+j1]*Mask[i1+1][j1+1]
            # print("{:4f}".format(x/6),end=" ");
            res[i, j] = round(x/6)
        print()
    return res
```

```python
res = conv(img, filter)
res
```

```python
def sort_m(A):
    tmp = np.zeros((len(A)+2, len(A[0])+2))
    for i in range(len(A)):
        for j in range(len(A[i])):
            tmp[i+1][j+1] = A[i][j]
    A = tmp
    for i in range(len(A)-2):
        for j in range(len(A[i])-2):
            res = []
            for i1 in range(-1,2):
                for j1 in range(-1,2):
                    res.append(A[i+1+i1][j+1+j1])
            res.sort()
            ind = 0
        for i1 in range(-1,2):
            for j1 in range(-1,2):
                tmp[i+1+i1][j+1+j1] = res[ind]
                ind+=1
    return tmp
  
```

```python
A_ex2 = [[5, 6, 9],
         [2, 5, 3],
         [8, 1, 2]]

A_ex2 = sort_m(A_ex2)
A_ex2
```

```python
M15 = cv2.boxFilter(I, -1, (15, 15), normalize = True)
M25 = cv2.boxFilter(I, -1, (25, 25), normalize = True)
(thres, K15) = cv2.threshold(M15, 125, 255, cv2.THRESH_BINARY)
(thres, K25) = cv2.threshold(M25, 125, 255, cv2.THRESH_BINARY)
A = [M15, M25, K15, K25]
```

```python
width=5
height=5
rows = 2
cols = 2
axes=[]
fig=plt.figure()

for a in range(rows*cols):
    axes.append( fig.add_subplot(rows, cols, a+1) )
    subplot_title=("Subplot"+str(a))
    axes[-1].set_title(subplot_title)  
    plt.imshow(A[a])
fig.tight_layout()    
plt.show()
```
