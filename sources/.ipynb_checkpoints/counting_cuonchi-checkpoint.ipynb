```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage.morphology import binary_fill_holes
import skimage.filters as filt
from skimage.segmentation import clear_border
from skimage.morphology import binary_opening, binary_closing, binary_erosion, binary_dilation, disk, dilation, erosion, remove_small_objects
```

```python
I = cv.imread('cuonchi.jpg')
gray_im = cv.cvtColor(I, cv.COLOR_BGR2GRAY)
plt.imshow(gray_im,cmap='gray')
```

```python
img_erose = erosion(gray_im, disk(6))
img_dilate = dilation(img_erose, disk(6))
plt.imshow(img_dilate,cmap='gray')
```

```python
img_sobel = filt.sobel(img_dilate)
# img_erose_2nd = erosion(img_sobel, disk(2))
# img_dilate_2nd = dilation(img_dilate, disk(2))
# print(np.maximum(img_sobel))
img_binary = img_sobel > 0.04
plt.imshow(img_binary,cmap='gray')
```

```python
fill_holes = binary_fill_holes(img_binary).astype(int)
plt.imshow(fill_holes, cmap='gray')
```

```python
im_dilate_3nd = erosion(fill_holes, disk(7))
print(type(im_dilate_3nd))
img_crop = im_dilate_3nd[180:-300,1200:-550]
plt.imshow(img_crop, cmap='gray')
```

```python
img_clear_bd = clear_border(img_crop)
plt.imshow(img_clear_bd, cmap='gray')
```

```python
img_dilate2nd = erosion(img_clear_bd, disk(15))
print(img_dilate2nd.shape)
plt.imshow(img_dilate2nd,cmap='gray')
```

```python
plt.imshow(gray_im,cmap='gray')
```

```python
img_complete = np.zeros(img_dilate2nd.shape)
for i in range(img_dilate2nd.shape[0]):
    for j in range(img_dilate2nd.shape[1]):
        if(img_dilate2nd[i][j]==0):
            img_complete[i][j]=0
        else:
            img_complete[i][j]=255
```

```python
print(img_complete[500:550,500:550])
img_complete = img_complete.astype('uint8')
print(img_complete[500:550,500:550])
print(img_complete.dtype)
```

```python
img_complete = dilation(img_complete, disk(18))
plt.imshow(img_complete, cmap='gray')
```

```python
# Labeling

ret, labels = cv.connectedComponents(img_complete)
label_hue = np.uint8(179 * labels / np.max(labels))
blank_ch = 255 * np.ones_like(label_hue)
labeled_img = cv.merge([label_hue, blank_ch, blank_ch])
labeled_img = cv.cvtColor(labeled_img, cv.COLOR_HSV2BGR)
labeled_img[label_hue == 0] = 0

# plt.subplot(222)
plt.title('Objects counted:'+ str(ret-1))
plt.imshow(labeled_img)
print('objects number is:', ret-1)
plt.show()
```
