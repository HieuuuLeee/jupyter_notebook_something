# 0. Bài toán: Từ một bức ảnh, Hãy vẽ ra một bức hình từ tấm hình đó

# 1. Chuẩn bị thư viện và ảnh

```python
#Khời tạo thư viện
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans
```

```python
I = mpimg.imread('halan.jpg')
```

```python
figure = plt.figure(figsize=(8,8))
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.tight_layout()
plt.imshow(I)
plt.title('Ảnh gốc'), 
plt.xticks([]) 
plt.yticks([])
plt.show()
```

# 2. Tạo bản vẽ phác thảo - bằng thuật toán tìm biên cạnh Canny

```python
edges = cv.Canny(I,70,200)
```

```python
figure = plt.figure(figsize=(15,15))
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.tight_layout()
plt.subplot(121),plt.imshow(I)
plt.title('Ảnh gốc'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Ảnh sau khi phác hoạ'), plt.xticks([]), plt.yticks([])
plt.show()
```

# 3. Tạo bản vẽ sinh màu - bằng thuật toán k-mean với theo màu sắc với thang đo là CIELAB

* Lý do sử dụng k-mean với mỗi nhóm màu vì nếu đây là bản vẽ tay thì nó sẽ không nhiều màu để vẽ, và k-mean theo màu sắc sẽ phân cụm các nhóm màu lại với nhau để chỉ có một lượng màu nhất định được sử dụng.
* Lý do sử dụng thang đo CIELAB vì theo bài báo thì thang đo này tốt hơn

```python
k = 15
LAB = cv.cvtColor(I, cv.COLOR_RGB2LAB)
Ireshape = LAB.reshape((LAB.shape[0] * LAB.shape[1], 3))
```

```python
kmeans = KMeans(n_clusters = k)
kmeans.fit(Ireshape)
```

```python
colors = kmeans.cluster_centers_
labels = kmeans.labels_[:, np.newaxis]
```

```python
Inew = np.ones(Ireshape.shape)
count=0
for label in labels:
    Inew[count]=colors[label]
    count+=1
Inew = Inew.reshape(LAB.shape).astype(np.uint8)
Inew = cv.cvtColor(Inew, cv.COLOR_LAB2RGB)
```

```python
figure = plt.figure(figsize=(15,15))
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.tight_layout()
plt.subplot(121),plt.imshow(I)
plt.title('Ảnh gốc'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Inew)
plt.title('Ảnh sinh màu'), plt.xticks([]), plt.yticks([])
plt.show()
```

# 4. Ghép ảnh - Kết hợp 2 bức ảnh trên để tạo ra

```python
level = 0.5
Ifinal = (Inew.T*(1-(edges/255)).T + level*edges.T).T.astype(np.uint8)
```

```python
figure = plt.figure(figsize=(15,15))
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.tight_layout()
plt.subplot(121),plt.imshow(I)
plt.title('Ảnh gốc'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Ifinal)
plt.title('Ảnh thành quả'), plt.xticks([]), plt.yticks([])
plt.show()
```
