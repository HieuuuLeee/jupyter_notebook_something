### Import libraries

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

### Bai 1: 

```python
A = np.array([[2, -7, 1], [-6, 2, -3], [4, 3, 2]])
B = np.array([[-1], [0], [2]])
```

```python
print("A inner prod B = \n{}".format(np.dot(A, B)))
```

### Bai 2: 

```python
ran = np.arange(0, 2, 0.1)
```

```python
f = []
for i in ran:
    f.append(pow(i, i * i) - pow(pow(i, i), i))
```

```python
x1 = ran
y1 = f
```

```python
plt.plot(x1, y1)
plt.show()
```

### Bai 3: 

```python
ran = np.arange(0, 10, 0.1)
```

```python
f2 = []
for i in ran:
    f2.append(1 - 2 * pow(math.e, -i) * math.sin(2 * math.pi - 35))
```

```python
x2 = ran
y2 = f2
```

```python
plt.plot(x2, y2)
plt.show()
```

### Bai 4: 

```python
list_sin = []
list_cos = []
list_tan = []
list_x = np.arange(0, 2 * math.pi, math.pi / 16)
```

```python
for x in list_x:
    list_sin.append(math.sin(x))
    list_cos.append(math.cos(x))
    list_tan.append(math.tan(x))
```

```python
print("List tan: {}".format(list_cos))
```

```python
plt.plot(list_x, list_cos)
plt.plot(list_x, list_sin)
plt.plot(list_x, list_tan)
plt.ylim((-5, 5))
plt.show()
```
